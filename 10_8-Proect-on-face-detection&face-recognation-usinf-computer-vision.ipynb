```python
def read_img(path):
    img = cv2.imread('./Img/') ## reading image
    (h,w) = img.shape[:2] #fetching height and weight
    width = 500  # hard coding width
    ratio = width / float(w)  # preparing a ration for height
    height = int(h*ratio) #generating new height 
    return cv2.resize(img,(width,height))  # returing the reshape image
```
