>> **Types of Cross Validation**\
>> **1. LOOCV - Leave One Out CV:**
>>   -  LOOCV is a type of cross-validation technique commonly used in machine learning and statistical modeling. 
>>   - LOOCV is a special case of k-fold cross-validation, where k is equal to the number of samples in the dataset. 
>>   - In LOOCV, the model is trained on all but one data point, and the model's performance is evaluated on the data point that was left out. 
>>   - This process is repeated for every data point in the dataset, and the performance is averaged to give an estimate of the model's generalization performance.
>> 
>>    - *Advantages of LOOCV:*
>>      1. LOOCV gives an unbiased estimate of the model's generalization performance since it uses all the available data for training and testing.
>>      2. LOOCV is particularly useful when the sample size is small since it maximizes the use of the available data.
>>      3. LOOCV can be used to evaluate the stability of the model since it provides a measure of how much the performance of the model varies with different training sets.
>>    - *Disadvantages of LOOCV:*
>>      1. LOOCV can be computationally expensive, especially when the sample size is large since it requires training and testing the model for every data point.
>>      2. LOOCV can produce high variance estimates of the model's generalization performance, especially when the dataset is small since the model is trained on almost the same data each time, except for one data point.
>>      3. LOOCV can be biased when the dataset is imbalanced since it uses all the available data for training and testing, which may result in overfitting to the majority class.

> - Leave-One-Out Cross-Validation (LOOCV): LOOCV is useful when we have a small dataset and want to maximize the use of our data for training and testing. LOOCV is also useful when we have a high-variance model and want to get the lowest possible bias in our estimate of model performance. 
> - For example, suppose we have a dataset of 100 samples and a binary classification problem. We can use LOOCV to repeatedly train and test our model on all samples except one, and then use the held-out sample for testing. This process is repeated for each sample in the dataset, resulting in 100 iterations of the training and testing process.

>> **2. K-Fold Cross validation**:
>> - K-fold cross-validation is a widely used cross-validation technique in machine learning and statistical modeling. It involves dividing the dataset into k subsets (or folds) of roughly equal size. The model is trained on k-1 folds and validated on the remaining fold, and this process is repeated k times, with each fold serving as the validation set once. The performance of the model is averaged across the k iterations to obtain an estimate of the model's generalization performance.
>> - **The k-fold cross-validation technique has several advantages over other cross-validation techniques, including:**
>> 1. It reduces the variance of the estimate of the model's generalization performance compared to Leave-One-Out cross-validation.
>> 2. It is computationally efficient since the model is trained and tested k times, rather than n times as in Leave-One-Out cross-validation.
>> 3. It provides a more realistic estimate of the model's performance since it evaluates the model on multiple, independent subsets of the data.
>> - **However, k-fold cross-validation also has some limitations and potential issues, including:**
>> 1. It can be sensitive to the value of k chosen, with higher values of k resulting in a larger bias but lower variance, and lower values of k resulting in a smaller bias but higher variance.
>>2. It assumes that the data is independently and identically distributed, which may not be the case in some real-world scenarios.
>>3. It may lead to overfitting if the model is tuned based on the performance on the validation set during each fold.

>> In summary, k-fold cross-validation is a popular cross-validation technique that provides a more reliable estimate of the model's generalization performance compared to Leave-One-Out cross-validation. However, it has some limitations that should be considered when selecting a cross-validation technique for a particular machine learning or statistical modeling task. The value of k should be chosen carefully, and other cross-validation techniques, such as stratified k-fold or nested cross-validation, may be more appropriate in some scenarios.

> - K-Fold Cross-Validation: K-fold CV is useful when we have a larger dataset and want to balance the trade-off between bias and variance in our estimate of model performance. K-fold CV is also useful when we have limited computational resources and cannot afford to use LOOCV. 
> - For example, suppose we have a dataset of 1000 samples and a binary classification problem. We can use 5-fold CV to split the dataset into 5 equal-sized folds, and then repeatedly train and test our model on 4 of the folds and use the remaining fold for testing. This process is repeated for each fold, resulting in 5 iterations of the training and testing process.

>> **3. Stratified Cross Validation**:
>> - Stratified cross-validation (CV) is a technique used in machine learning to evaluate the performance of a model when the data is imbalanced, i.e., when the proportion of samples in each class is not equal. In stratified CV, the dataset is divided into K-folds, just like in regular K-fold CV, but the difference is that each fold maintains the same class distribution as the original dataset.
>> - The process of creating the folds in stratified CV involves partitioning the data into subsets, such that each subset has an equal proportion of samples from each class. This ensures that each fold has a representative distribution of samples from each class, which is important for accurate evaluation of the model's performance.
>> - Stratified CV is particularly useful when working with classification problems, where the goal is to predict the class of a sample based on its features. It helps to ensure that the model is able to generalize well to new, unseen data, especially when the classes are imbalanced.
>> - **Overall, the advantages of using stratified CV include:**
>>      -    It helps to ensure that the model is evaluated on a representative sample of the data, which is important when working with imbalanced datasets.
>>      - It provides a more accurate estimate of the model's performance, as it reduces the variance in the evaluation metric compared to regular K-fold CV.
>>      - It can help to identify potential problems with the model, such as overfitting or underfitting, by providing a more reliable estimate of the model's performance on new, unseen data.
>> - The main disadvantage of stratified CV is that it can be computationally expensive, especially when working with large datasets or complex models. Additionally, it may not always be necessary or useful to use stratified CV, especially when the classes are well-balanced.
>> - 

> - Stratified Cross-Validation: Stratified CV is useful when we have imbalanced classes in our dataset and want to ensure that each fold has a representative sample of each class. Stratified CV is also useful when we have a classification problem and want to estimate the model's performance on each class separately. 
> - For example, suppose we have a dataset of 1000 samples, of which 800 are negative examples and 200 are positive examples, and a binary classification problem. We can use 5-fold stratified CV to split the dataset into 5 equal-sized folds, such that each fold has 80% negative examples and 20% positive examples. We can then repeatedly train and test our model on each fold, and evaluate its performance on both the negative and positive examples separately.


>> **4.Time Series Cross Validation**:
>> - Time Series Cross-Validation is a type of cross-validation technique that is specifically designed for time series data. In traditional cross-validation techniques, the data is randomly split into training and test sets, with no consideration for the order or time dependency of the data. However, in time series data, the order of the data is important, and the training set should always come before the test set in time.
>> - Time Series Cross-Validation addresses this issue by splitting the data into multiple folds in a way that preserves the temporal order of the data. The data is split into a training set and a test set, with the test set being a contiguous block of time that comes after the training set. This process is repeated for each fold, with each fold containing a different contiguous block of time.
>> - One common approach to Time Series Cross-Validation is called "rolling window" or "expanding window" cross-validation. In this approach, the training set starts with a small window of time and gradually expands to include more and more historical data. The test set is always a fixed length of time that comes after the training set.
>> - For example, suppose we have a time series dataset with 1000 time points, and we want to use Time Series Cross-Validation with a rolling window approach and 5 folds. We could split the data as follows:

    Fold 1: Training set from time point 1 to 200, test set from time point 201 to 250
    Fold 2: Training set from time point 1 to 400, test set from time point 401 to 450
    Fold 3: Training set from time point 1 to 600, test set from time point 601 to 650
    Fold 4: Training set from time point 1 to 800, test set from time point 801 to 850
    Fold 5: Training set from time point 1 to 1000, test set from time point 951 to 1000
>> - Time Series Cross-Validation allows us to evaluate the performance of our model on future time points, which is important in time series forecasting. It also helps to prevent overfitting by ensuring that the model is not using future data to train, which could result in overly optimistic performance estimates.

### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

### Importing Data

Link : https://raw.githubusercontent.com/krishnaik06/Types-Of-Cross-Validation/main/cancer_dataset.csv

```python
df = pd.read_csv ("https://raw.githubusercontent.com/krishnaik06/Types-Of-Cross-Validation/main/cancer_dataset.csv")
```

```python
df.head()
```

```python
df.drop("Unnamed: 32", axis = 1, inplace = True)
```

```python
X = df.iloc[:,2:]
y = df.iloc[:,1]
```

```python
X
```

```python
y
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

for i in range(100):
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.5, random_state=i)
    model = LogisticRegression()
    model.fit(X_train,y_train)
    result = model.score(X_test, y_test)
    print(result)

# When random state == 0
# 0.9543859649122807

# When random state == 100
# 0.9438596491228071

# Accuracy is dependent on random state.
# As we can see that using random_state the score is getting fluctuate, so it will not be easy to 
# tell the accuracy of the model while we are changing the random state so it is difficult the 
# predict the accuracy.
```
