> # **Decision Tree Classification**
> It can use in classification as well as Regression.
> - Decision Tree Classification is a machine learning algorithm that uses a tree-like model to make decisions based on a set of features. 
> - It works by recursively partitioning the data into subsets based on the values of the input features, and creating a tree-like structure where each node represents a feature and each branch represents a decision based on the value of that feature. 
> - The algorithm learns from training data and uses the tree to classify new data by following the path of the tree based on the values of the input features, until a final decision is made at a leaf node. 
> - Decision Tree Classification is often used for problems where there are multiple potential decision paths based on different features, and is known for its ability to handle both numerical and categorical data.
>
> **Pros:**
> 1. Easy to use and Interpret.
> 2. Can be visualized
> 3. Little data preparation (non-parametric [the data whis is not normal])
> 4. Able to handle bothe numerical and categorical data.
> 5. Multi-output problems can be handled.
> 6. Use boolean logic to model the predictions.
> 7. Performs well (some of the assumptions)
>
> **Cons:**
> 1. Comples tree for Beginners
> 2. Unstable
> 3. No smooth or continuous predictions.
> 4. Complex steps (XOR, partly or multiplexer problem)
> 5. Decision tree learners create biased trees if some classes dominate.
>

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier

X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(8,8))
model = DecisionTreeClassifier().fit(X,y)
plot_tree(model, filled=True)

plt.title("Decission Tree Trained Model of IRIS Data")
plt.show()
```

>> ## **Gini index, or Gini impurity**
>>- The Gini index, or Gini impurity, is a measure of the impurity or uncertainty of a classification decision made by a decision tree algorithm. 
>>- In simple terms, it measures how often a randomly chosen element in a dataset would be incorrectly classified based on the decision tree model. 
>>- It ranges between 0 and 1, where 0 indicates that all elements belong to the same class and 1 indicates that elements are equally distributed across all classes.
>>- A lower Gini index indicates a more "pure" node in the decision tree, with a higher probability of correctly classifying an element. 
>>- The Gini index is commonly used in decision tree-based algorithms, including the popular CART (Classification and Regression Trees) algorithm.

> # Random Forest Classifier
>- Random forest classification is a machine learning algorithm used for both classification and regression problems. 
>- It is an ensemble (a group of items viewed as a whole rather than individually) learning method that creates multiple decision trees and combines their predictions to make the final decision. 
>- Each decision tree is constructed using a random subset of the training data and a random subset of the features. 
>- Random forest classification algorithm is effective in reducing the overfitting problem associated with decision tree algorithms. The output of the algorithm is the most common prediction made by all the decision trees in the forest.
