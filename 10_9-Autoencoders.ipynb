## Autoencoders

- It is type of ANN which is used to learn data encoding in and unsupervised manner.



|Architecture of Autoencoders| |
|---|---|
|1.Encoder| It is the set of Convolutional Blocks(Two types Pooling and Convolutional)|
|2.Decoder| It is the set of Convolutional Blocks but also the set of upsampling. It act as an encoder but in reverse manner. Beacause now we dont have to shrink the image nor we don't have to compress the image. We are Decoding the image so we have to use the Reverse Convolutional Block|
|3.Bottle Neck| |

- Parameters:
    1. BottleaNeck Size: It decides how much data has to be compressed.
    2. Number of layers: Number of hidden layers.
    3. Number of neurons per layer.

- Autoencoders types:
  1. Unddrcompletw
  2. Sparse
  3. Contractive
  4. Denoising
