> **CNN --> Convolutional Neural Network**
> - CNN, or Convolutional Neural Network, is a type of neural network that is commonly used in image and video recognition tasks. The name "convolutional" comes from the mathematical operation used in the network, called convolution.
>
>  - The key feature of CNNs is that they are able to automatically learn and identify features in an image, such as edges, corners, and patterns, through a process called convolution. This is done by applying a set of filters or kernels to the input image, which scan the image pixel by pixel and detect important features.
>
> - The output of this convolutional layer is then passed through one or more pooling layers, which downsample the image by taking the maximum or average value of adjacent pixels. This helps to reduce the size of the image and the number of parameters in the network, making it faster and more efficient.
>
> - The final layers of a CNN are typically fully connected layers, which are similar to those found in traditional neural networks. These layers use the features extracted by the convolutional and pooling layers to make a prediction or classification.
>
> - Overall, CNNs are very powerful and effective in image and video recognition tasks because they are able to automatically learn and extract important features from the input data, without the need for manual feature engineering.

>- The Rectified Linear Unit (ReLU) activation function is commonly used in neural networks because of its simplicity and effectiveness. Here are some reasons why ReLU is a popular choice:
>   1. Non-linearity: The ReLU function introduces non-linearity into the neural network. This allows the network to learn more complex and nuanced relationships between inputs and outputs, which is essential for many machine learning tasks.
>   2. Sparsity: The ReLU function has a sparsity property, which means that it outputs zero for negative inputs. This sparsity property can help to reduce the likelihood of overfitting by preventing the network from relying too heavily on any one input or feature.
>   3. Computationally efficient: The ReLU function is computationally efficient to evaluate, making it well-suited for large-scale neural networks that require many calculations.
>   4. Avoids vanishing gradient: The ReLU function avoids the vanishing gradient problem that can occur with other activation functions, such as the sigmoid function. The vanishing gradient problem occurs when the gradients become very small during backpropagation, which can make it difficult to train deep neural networks.
>- Overall, the ReLU function is a simple and effective activation function that is well-suited for many machine learning tasks.


>> **1. Convolutional (Filter/Kernal)**
>>
>> In the context of CNNs, a convolutional filter/kernel is a small matrix of numbers that is used to extract specific features from an input image. The filter is a sliding window that moves over the input image pixel by pixel, and at each position, the values of the filter are multiplied by the corresponding pixel values of the input image.
>>
>>The resulting products are then summed up to produce a single output value, which represents the presence or absence of a particular feature in the input image at that location.
>>
>>The filter/kernel is typically designed to detect specific features, such as edges or textures, by emphasizing certain patterns of pixel values and de-emphasizing others.
>>
>>By using multiple filters with different patterns, a CNN can extract multiple features from an input image and use them to classify or recognize the image.

>> **2. Pooling Max/Avg**
>>
>>- It will going to perform the process of extraction of perticular vlaues from the sets of values, usually the max value or avg value.
>>- The motive is to reduce the size of the output matrix.
>>- Pooling is a common operation in convolutional neural networks (CNNs) used to reduce the spatial dimensions of the input data by summarizing its features. Max pooling and average pooling are two common types of pooling operations.
>>- In max pooling, the input data is divided into non-overlapping rectangular blocks, and the maximum value within each block is retained while discarding the other values. This operation effectively reduces the dimensionality of the input data and focuses on the most important features. For example, in an image, a max pooling operation on a block of pixels would retain the pixel with the highest intensity while ignoring the others.
>>- In average pooling, the input data is divided into blocks and the average value within each block is calculated. This operation can also help reduce the dimensionality of the input data, but it retains more information than max pooling.
>>- Both max pooling and average pooling are used in CNNs to help reduce the spatial dimensions of the input data, allowing the network to focus on the most important features while reducing the computational complexity.
