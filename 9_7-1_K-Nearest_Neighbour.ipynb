## K Nearest Neighbour

> - It is a simple Classification algorithem.
> 
> 
> 
> - The KNN algorithm works by finding the K nearest neighbors to a given data point based on a distance metric, such as Euclidean distance. 
> 
> - The value of K is a hyperparameter that must be specified before the algorithm is trained. To make a prediction for a new data point, the algorithm looks at the K nearest neighbors to that data point in the training data and assigns the new data point to the most common class among those K neighbors.
> - The KNN algorithm can be useful in situations where the decision boundary between classes is highly non-linear and cannot be easily modeled using a linear classifier. 
> - However, it can be sensitive to the choice of K and the distance metric used, and can be computationally expensive for large datasets or high-dimensional feature spaces.

>- Distance methods are techniques used in machine learning for measuring the similarity or dissimilarity between data points in a dataset. The goal of using distance measures is to cluster similar data points together and separate dissimilar data points.
>- There are different distance methods used in ML, such as **Euclidean distance, Manhattan distance, and cosine distance**. 
>- **Euclidean distance** calculates the straight-line distance between two points in space, while **Manhattan distance** calculates the distance between two points along the axes of a grid or a graph. 
>- **Cosine distance** calculates the angle between two vectors in a multidimensional space. 
>- These distance methods are used in various algorithms, such as k-Nearest Neighbors, Hierarchical Clustering, and Principal Component Analysis, among others.
>>- Distance methods are widely used in machine learning for various tasks such as clustering, classification, and anomaly detection.
>>-Here are a few examples of when and where distance methods can be used:
>>1. Clustering: In clustering, we group similar data points together. Distance methods such as k-means and hierarchical clustering use distance measures to group similar data points.
>>2. Classification: In classification, we predict the class of a new data point based on its features. Distance methods such as k-nearest neighbors (k-NN) use distance measures to find the k closest neighbors and predict the class based on the majority class of the neighbors.
>>3. Anomaly detection: In anomaly detection, we identify data points that are significantly different from the rest of the data. Distance methods such as local outlier factor (LOF) use distance measures to identify data points that are far from their neighbors.
>>- Overall, distance methods are useful in machine learning when we need to measure the similarity or dissimilarity between data points.
