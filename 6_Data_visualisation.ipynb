## Labels

```python
x = [3,8,1,10]
y = [6,2,7,11]

plt.plot(x);
plt.plot(y);

# for labels
plt.xlabel ("Height")
plt.ylabel ("Weight")
```

### Title

```python
plt.plot(x);
plt.plot(y);

# for labels
plt.xlabel ("Height")
plt.ylabel ("Weight")

# for title
plt.title ("Height v/s Weight Graph");
```

- modification can be use in xlabel, ylabel, and in title too no need to do more changes in it.

```python
plt.plot(x);
plt.plot(y);

# for labels
plt.xlabel ("Height")
plt.ylabel ("Weight")

modification = {'color':'red', 'size' : 15, 'family': 'serif'}

# for title
plt.title ("Height v/s Weight Graph", fontdict = modification);
```

#### If we want to change the title position.

- use to function loc (location)

```python
plt.plot(x);
plt.plot(y);

# for labels
plt.xlabel ("Height")
plt.ylabel ("Weight")

modification = {'color':'red', 'size' : 15, 'family': 'serif'}

# for title
plt.title ("Height v/s Weight Graph", fontdict = modification, loc = 'right');
# right
# center
```

## Subplots
- If we want our two data in single canvas we can use the subplot
  - plt.subplot (row, columns, plot_number)

```python
# 1st plot

x = [1,2,3,4,5,6]
y = [10,20,30,40,50,60]

plt.subplot (1,2,1)
plt.plot(x,y);
# 2nd plot
x = [3,8,1,10]
y = [6,2,7,11]

plt.subplot (1,2,2)
plt.plot(x,y);

plt.suptitle("Animals");
```

## Ticks and Ticks Labels

```python
x = [3,8,1,10]
y = [6,2,7,11]

# to change the default size of figure use add_axes function
fig = plt.figure()
ax = fig.add_axes([1,1,2,2])

# After changing the figure size use ax as varaible insted of plt

ax.plot(x,y);
# for title after changing the figure size use following functions for labels, titles 
ax.set_title("ANIMALS")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")

#  To change the X Tick Labels
# ax.set_xticks([10,20,30,40,50]);

# If we want to keep our graphical data at same position and just want to change the x tick lables use the following function
ax.set_xticklabels([101,102,103,104,105,106,107]);


#  To change the Y Tick Labels
# ax.set_yticks([3,6,9,12,15,18]);
# Similarly for the Y axis
ax.set_yticklabels(['One','Two','Three','Four','Five','Six','Seven']);
```

## Scatter plot:

- 0.5 < r < 0.75 ---> Slight Positive Correlation
- 0.75 < r < 0.99 ---> Strong Positive Correlation
- r == 1 ---> Perfect Positive Correlation

```python
x = [25,25,35,15]

plt.pie(x,labels = ['choc','vina','straw','blackcurrent'],
        startangle = 90, explode = [0.2,0,0.1,0],
        shadow = True, colors = ['g','y','k','blue']);

# Legends convay what exactly those color meant
plt.legend(title = "Ice Cream flavors");
```
