>> ## Simple Linear Regression
>> - When you want to find out the relationship b/w one independent varaible(one feature) with respect to one dependent variable (one > target)
>> - Simple linear regression is a statistical method used to identify the relationship between two continuous variables by fitting a straight line to the data. One variable is considered the independent variable, while the other is the dependent variable. 
>> - The aim of simple linear regression is to find the best fit line that minimizes the sum of the squared errors between the predicted values and the actual values. 
>> - The equation of the line can be used to predict the value of the dependent variable for any given value of the independent variable. This method is often used in data analysis to determine the strength and direction of the relationship between two variables.
>> 
>> 
> 
>> ## Multiple Linear Regression
>> - When we want to find out the relationship b/w multiple independent variable with respect to one dependent variable.
>> - Multiple linear regression is a statistical technique that is used to analyze the relationship between two or more independent variables and a dependent variable. It is an extension of simple linear regression, which only involves one independent variable.
>> - In multiple linear regression, a regression equation is developed to predict the value of the dependent variable based on the values of two or more independent variables. The goal is to find the best linear combination of the independent variables that can predict the value of the dependent variable as accurately as possible.

## Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Assign the dataset to the DataFrame

```python
var_x = np.array([1.1, 1.3, 1.5, 2.0, 2.2, 2.9, 3.0, 3.2, 3.2, 3.7, 3.9, 4.0, 4.0, 4.1, 4.5, 4.9, 5.1, 5.9, 5.3, 6.0, 6.1, 7.9, 8.2, 8.7, 9.0, 9.5, 9.6, 10.3, 10.5, 6.8, 7])
var_y = np.array([39343, 46205, 37731, 43535, 39821, 56642, 60150, 54445, 64445, 57189, 63218, 55794, 56957, 57081, 61111, 67938, 73504, 79123, 83088, 81363, 93940, 91738, 98217, 101302, 113812, 109431, 105582, 116969, 12635, 122391, 121872])
```

```python
len(var_x)
```

```python
len(var_y)
```

```python
plt.scatter(var_x,var_y);
```

Create a Dataframe containing the two columns "Experience" and "Salary" from the above the data.

```python
df = pd.DataFrame({'experience': var_x, "salary":var_y})
df.head()
```

```python
df.info()
```

## Form a variable "X" to store all the features and "y" to store the targets

```python
X = df.experience.values.reshape(-1, 1)
y = df.salary
```

## Splitting the data into two sets - Training set and Testing sets

```python
import sklearn
from sklearn.model_selection import train_test_split
```

### For Hyper-Parameter tuning to increase more accuracy of the model.
random_state = depends on len of df

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state= 1)
```

```python
len(X_train)
```

```python
len(X_test)
```

## Choose the model

```python
from sklearn.linear_model import LinearRegression 
```

```python
model = LinearRegression()
```

## Train the model on the trainig set

```python
model.fit(X_train, y_train)
```

## Make Predictions

```python
y_test
```

```python
y_pred = model.predict(X_test)
```

## Performance Metric - MSE

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_pred)
```

## Accuracy

```python
model.score(X_test,y_test)
```

## R^2

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```
