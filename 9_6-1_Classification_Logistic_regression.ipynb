- Logistic Regression is a statistical method used to analyze and model relationships between a dependent variable (usually binary or categorical) and one or more independent variables (which can be categorical or continuous).
- It is called "logistic" because it uses a logistic function, which is an S-shaped curve, to model the relationship between the independent variables and the probability of the dependent variable being a certain category.
- In simpler terms, logistic regression is used to predict the probability of a particular event occurring, based on the values of the input variables.
- For example, it can be used to predict the likelihood of a customer buying a product, based on their age, gender, income, and other factors.
- The output of logistic regression is a probability value between 0 and 1, which represents the likelihood of the event occurring. This probability value can be converted into a binary outcome by setting a threshold value (e.g., 0.5), above which the event is considered to occur, and below which it is considered not to occur.
- Logistic regression can also be used to estimate the impact of each independent variable on the dependent variable. This is done by calculating the odds ratio, which is the ratio of the probability of the event occurring for one group (e.g., those with a certain level of income) to the probability of the event occurring for another group (e.g., those with a different level of income).
- Overall, logistic regression is a powerful and widely-used method for modeling and analyzing binary or categorical outcomes. It can be applied in various fields, such as marketing, finance, healthcare, and social sciences, to make predictions and gain insights from data.

- Normalization (MinMax Scaler)
  -  Normalization is a technique used in machine learning to rescale the features or variables of a dataset to a similar range or scale, usually between 0 and 1. This is done to ensure that each feature contributes equally to the analysis and to prevent features with larger values from dominating the model.
  -  Normalization is particularly useful when working with features that have different units of measurement or scales, or when the features have a wide range of values. It can improve the performance of some machine learning algorithms, such as K-nearest neighbors and neural networks, by reducing the influence of outliers and improving the convergence of the optimization algorithms.
     -  Normalization can be done using various techniques, such as Min-Max scaling, Z-score normalization, and L1 and L2 normalization
     -  Min-Max scaling rescales the feature values to a range between 0 and 1, by subtracting the minimum value and dividing by the range.
     -  Z-score normalization rescales the feature values to have a mean of 0 and a standard deviation of 1, by subtracting the mean and dividing by the standard deviation.
     - L1 and L2 normalization rescale the feature values based on their magnitude, by dividing each value by the L1 or L2 norm of the feature vector.
  - Normalization should be used when the scale and range of the features can affect the performance of the machine learning model. For example, in image classification, normalization can be used to rescale the pixel values of an image to a similar range, which can improve the accuracy of the model. In text classification, normalization can be used to rescale the frequency of word occurrences to a similar range, which can improve the performance of the model.
  - Overall, normalization is a useful technique in machine learning for rescaling the features or variables of a dataset to a similar scale, which can improve the performance of some machine learning algorithms and prevent features with larger values from dominating the analysis.
